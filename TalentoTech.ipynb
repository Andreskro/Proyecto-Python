```python
# prompt: ejemplos con while

# Ejemplo 1: Contar hasta un número dado

numero = int(input("Ingrese un número: "))
contador = 0

while contador <= numero:
  print(contador)
  contador += 1

# Ejemplo 2: Sumar números hasta que el usuario ingrese 0

suma = 0
numero = int(input("Ingrese un número (0 para terminar): "))

while numero != 0:
  suma += numero
  numero = int(input("Ingrese un número (0 para terminar): "))

print("La suma de los números ingresados es:", suma)


# Ejemplo 3: Validación de entrada

while True:
  edad = input("Ingrese su edad: ")
  if edad.isdigit():
    edad = int(edad)
    if 0 < edad < 120:  # Ejemplo de rango válido
        print("Edad válida:", edad)
        break
    else:
        print("Edad fuera de rango.")
  else:
    print("Entrada inválida. Ingrese un número.")

# Ejemplo 4: Bucle infinito con condición de salida

continuar = True
while continuar:
    respuesta = input("¿Desea continuar? (s/n): ")
    if respuesta.lower() == "n":
        continuar = False
    else:
        print("Continuando...")

print("Fin del programa.")


# Ejemplo 5:  Mostrar los números pares hasta un limite

limite = int(input("Ingrese un limite: "))
numero = 0
while numero <= limite:
    if numero % 2 == 0:
        print(numero)
    numero += 1
```

```python
# Inicialización de contadores
negativos = 0
positivos = 0
neutros = 0

# Realizar 6 iteraciones
for i in range(0,6):
    # Pedir al usuario que ingrese un número
    numero = int(input(f"Ingrese el número {i+1}: "))


    if numero < 0:
        negativos += 1
    elif numero > 0:
        positivos += 1
    else:
        neutros += 1

# Imprimir los resultados
print(f"Cantidad de números negativos: {negativos}")
print(f"Cantidad de números positivos: {positivos}")
print(f"Cantidad de números neutros (cero): {neutros}")
```

# Objetivo
Realizar un analisis exploratorio de datos sobre la transicion energetica en colombia, examinando la evolucion de las fuentes de energia  el consumo a lo largo del tiempo
# Datos ficticios

Vamos a utilizar un conjunto de datos ficticios en formato CSV con la siguiente estructura.

* año : Año del registro  
* fuente_energia :  Tipo de fuente de energia (solar, eolica, etc)  
* generacion_mwh : Generacion total MWh (mega vatios hora) de esa fuente  
* consumo_mwh: Consumo total en MWh en el pais al año


# Pasos del ejercicio
1. Importar librerias y datos
2. Explorar el conjunto de datos
3. Analasis de tendencias de generacion de energia
4. Comparar la genracion por fuente de energia
5. Visualizacion de resultados










```python
# Paso 1: Importar las librerias y datos
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# cargar datos ficticios

data= {
    'año': [2018, 2018, 2018, 2019, 2019, 2019, 2020, 2020, 2020],
    'fuente_energia': ['solar', 'eolica', 'hidroelectrica', 'solar', 'eolica', 'hidroelectrica', 'solar', 'eolica', 'hidroelectrica'],
    'generacion_mwh': [500, 300, 1000, 600, 350, 1100, 700, 400, 1200],
    'consumo_mwh': [1500, 1500, 1500, 1600, 1600, 1600, 1700, 1700, 1700]
}

df = pd.DataFrame(data)

#paso 2: Explorar el Conjunto de datos
print(df.head())
print(df.describe())
print(df.info())


```

```python
generacion_tenedencia = df.groupby('año')['generacion_mwh'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(generacion_tenedencia['año'], generacion_tenedencia['generacion_mwh'], marker= 'o')
plt.plot(data['año'], data['generacion_mwh'], marker= 'o')
plt.title('Tendencia de generacion de energia en Colombia')
plt.xlabel('Año')
plt.ylabel('Generacion total (MWh)')
plt.grid(True)
plt.show()
```

```python
generacion_tenedencia
```

```python
#Grafica en consumo
tendencia_consumo = df.groupby('año')['consumo_mwh'].unique().reset_index()
print(tendencia_consumo)
plt.figure(figsize=(10, 8))
plt.plot(tendencia_consumo['año'], tendencia_consumo['consumo_mwh'], marker= 'o')
plt.title('Tendencia de consumo de energia en Colombia')
plt.xlabel('Año')
plt.ylabel('Consumo total (MWh)')
plt.grid(True)
plt.show()
```

```python
# comparar la generacion por funete de energia

generacion_por_fuente = df.groupby(['año', 'fuente_energia'])['generacion_mwh'].sum().unstack()

generacion_por_fuente.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title('Generacion por fuente de energia en Colombia')
plt.xlabel('Año')
plt.ylabel('Generacion total (MWh)')
plt.legend(title='Fuente de energia')
plt.show()
```

# Problema: Análisis de datos meteorológicos

Supongamos que tenemos un conjunto de datos meteorológicos que contiene la temperatura diaria máxima en grados Celsius registrada en diferentes ciudades durante un año. El objetivo es realizar las siguientes operaciones utilizando NumPy:

Encontrar la temperatura máxima registrada en todas las ciudades durante el año.
Calcular la temperatura promedio diaria para cada ciudad.
Identificar los días en los que la temperatura superó los 30 grados Celsius en al menos una ciudad.
Calcular la cantidad de días de lluvia, donde se considera un día de lluvia si la temperatura máxima es inferior a 10 grados Celsius.
Asumiendo que los datos están organizados en un array NumPy llamado temperaturas donde cada fila representa un día y cada columna representa una ciudad diferente.

```python
import numpy as np

temperaturas =np.array([[26, 28, 30, 29, 27],
                        [31, 32, 33, 30, 29],
                        [29, 27, 28, 31, 30],
                        [32,35,36,23,32],
                        [8,12,15,4,8],
                        [36,39,25,28,32],
                        [23,24,25,35,32],
                        [25,34,28,25,22],
                        [12,11,14,25,15],
                        [26,27,30,31,29],
                        [12,8,12,6,9],
                        [25,21,27,21,28]])
```

```python
#Encontrar la temperatura maxima registrada en todas las ciudades durante el año
temp_max_global=np.max(temperaturas)
print("Temperatura maxima registrada en todas las ciudades durante el año: ", temp_max_global)

#Calcular la temperatura promedio diaria para cada ciudad
temp_promedio_ciudad=np.mean(temperaturas, axis=0)
print("Temperatura promedio diaria para cada ciudad: ", temp_promedio_ciudad)

#Identificar los dias en lo que temperatura supero los 30 grados celsius en al menos una ciudad
dias_calurosos=np.any(temperaturas>30, axis=1)
print("Dias en los que la temperatura supero los 30 grados Celsius: ", np.sum(dias_calurosos))

#Identificar la cantidad de dias de lluvia
dias_de_lluvia=np.sum(temperaturas<10)
print("Cantidad de dias de lluvia: ", dias_de_lluvia)
```

```python
anomalias=[]
#creamos una funcion que, encuentra los atipico o outliers
def buscar_anomalias(dataframe):
  df_std=dataframe.std()
  df_mean=dataframe.mean()
  corteAnomalo=df_std*2
  limite_menor=df_mean-corteAnomalo
  limite_mayor=df_mean+corteAnomalo
  print(limite_menor.iloc[0])
  print(limite_mayor.iloc[0])

  #generacion de las anomalias
  for index, row in dataframe.iterrows():
    outlier=row # obtener columna    fila: columna1|columna2|columna3
    if (outlier.iloc[0]>limite_mayor.iloc[0]) or (outlier.iloc[0]<limite_menor.iloc[0]):
      anomalias.append(index)
  return anomalias

resultado_anomalias = buscar_anomalias(df_esp.set_index("alpha_3")[["population"]])
print(resultado_anomalias)

```

```python
import matplotlib.pyplot as plt
import pandas as pd
df['Date'] = pd.to_datetime(df['Date'])

data = df.dropna(subset=['Date'])

data['Month'] = data['Date'].dt.month

monthly_counts = data['Month'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
monthly_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Cantidad de accidentes por mes', fontsize=16)
plt.xlabel('Mes', fontsize=14)
plt.ylabel('Cantidad de accidentes', fontsize=14)
plt.xticks(ticks=range(12), labels=['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
                                    'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'], rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
Mayo = monthly_counts.idxmax()
Julio= monthly_counts.idxmin()

max_min_data = monthly_counts.loc[[Mayo, Julio]]

plt.figure(figsize=(8, 5))
max_min_data.plot(kind='bar', color=['blue', 'green'], edgecolor='grey')
plt.title('Comparación entre el mes con más y menos accidentes', fontsize=16)
plt.xlabel('Mes', fontsize=14)
plt.ylabel('Cantidad de accidentes', fontsize=14)
plt.xticks(ticks=[0, 1], labels=['Mayo', 'Julio'], rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
